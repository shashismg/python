```python
df_xl_movies = pd.read_excel("movies_db.xlsx", "movies")
df_xl_movies.head(3)
```

```python
df_xl_fin = pd.read_excel("movies_db.xlsx", "financials")
df_xl_fin.head(3)
```

```python
df_merged = pd.merge(df_xl_movies, df_xl_fin, on='movie_id')
df_merged.head(3)
```

```python
df_merged.to_excel("movies_merged.xlsx", sheet_name="merged", index=False)
```
